1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Визначаємо символічну змінну
x = sp.symbols('x')

#Задаємо лінійну функцію, беремо похідну
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)
print('Похідна лінійної функції 10x-5:', derivative_f_lin)
```

```python
#Задаємо лінійний простір
x = np.linspace(0, 10, 100)
f_lin_func = 10 * x - 5
df_lin = [10] * 100

#Створення графіку
plt.figure(figsize=(10,5))
plt.plot(x, f_lin_func, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df_lin, label="f'(x) = 10", color='green')
plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)
print('Похідна тригонометричної функції sin(x):', derivative_f_sin)
```

```python
#Задаємо лінійний простір
x = np.linspace(0, 10, 100)
f_sin_func = np.sin(x)
df_sin = np.cos(x)

#Створення графіку
plt.figure(figsize=(10,5))
plt.plot(x, f_sin_func, label='f(x) = sin(x)', color='blue')
plt.plot(x, df_sin, label="f'(x) = cos(x)", color='green')
plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```

```python
x = sp.symbols('x')
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)
print('Похідна логарифмічної функції ln(x):', derivative_f_ln)
```

```python
#Задаємо лінійний простір
x = np.linspace(0, 10, 100)
f_ln_func = np.log(x)
df_ln = 1/x

#Створення графіку
plt.figure(figsize=(10,5))
plt.plot(x, f_ln_func, label='f(x) = ln(x)', color='blue')
plt.plot(x, df_ln, label="f'(x) = 1/x", color='green')
plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
derivative_f = sp.diff(f, x)
print('Похідна квадратичної функції x^2 - 5x + 1:', derivative_f)
```

```python
#Задаємо лінійний простір
x = np.linspace(0, 100, 100)
f = x**2 - 5*x + 1
df = 2*x - 5

#Створення графіку
plt.figure(figsize=(10,5))
plt.plot(x, f, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green')
plt.title('Графік функції та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```
